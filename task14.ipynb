```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")
```

x ~ N(a, σ_x²=2)
y ~ N(b, σ_y²=1)

H0: a=b
H1: a>b

(< x > - a)*sqrt(n)/σ_x ~ N(0, 1)
(< y > - b)*sqrt(n)/σ_y ~ N(0, 1)

Тогда < x > - a - (< y > - b) ~ N(0, σ_x²/n + σ_y²/m)

Если a=b (H0), то < x > - < y > ~ N(0, σ_x²/n + σ_y²/m)

```python
α = 0.05

n = 3
m = 2

x_n = np.array([-1.11, -6.10, 2.42])
y_m = np.array([-2.29, -2.91])

σ2_x = 2
σ2_y = 1

Δ = x_n.mean() - y_m.mean()
print(f"Δ = {Δ:.4f}")
N = st.norm(0, np.sqrt(σ2_x / n + σ2_y / m))
U = N.ppf(1 - α)

print(f"Поскольку a>b то рассматриваем квантиль 1-α: {U:.4f}")

p_value = N.sf(Δ)
print(f"p-value = {p_value:.4f}, значит нет веских оcнований отвергнуть H0")
```

Пусть a-b=δ, тогда Δ = < x > - < y > - δ ~ N(0, σ_x²/n + σ_y²/m)

< x > - < y > ~ N(δ, σ_x²/n + σ_y²/m)

Следовательно W(δ) = P(Δ > U_(1-α))

```python
plt.figure(figsize=(12, 8))

δ = np.linspace(0, 6, 100)
W = N.sf(-δ + U)
plt.plot(δ, W, linewidth=2.5)


plt.xlabel('δ', fontsize=14)
plt.ylabel('W(δ)', fontsize=14)
plt.title('Мощность критерия')
plt.tight_layout()
plt.show()
```
